**Breve explicación del algoritmo simmulated annealing (1 minuto)**

Hacer una explicación breve del algoritmo.

"
*El algoritmo de recocido simulado, SA por sus siglas en inglés (Simulated Annealing), es un algoritmo de búsqueda metaheurística para problemas de optimización GLOBAL. 
Está inspirado en el proceso de recocido del metal que consiste en calentar y enfriar el material para poder manipular el estado y el desplazamiento de los átomos y así alterar las propiedades físicas de este.
Teniendo una función $f$, el algoritmo consiste en tomar un punto $x_a$ de manera aleatoria para obtener $f(x_a)$ y, escogiendo aleatoriamente un vecino $x_b$, decidir por probabilidad si permanecer en el punto $x_a$ o cambiar a $x_b$ con el fin de llegar a un punto máximo o mínimo global de la función.*
"

**Profundización en el algoritmo. (1 minuto)**

Explicar la dinámica del algoritmo, como funciona.

"
*Esta probabilidad está sujeta a una variable $T$ (temperatura) que va disminuyendo en cada iteración del algoritmo. Es importante aclarar que el algoritmo no siempre converge al punto óptimo.
Supongamos que queremos hallar el mínimo de la función.
Siempre que $f(x_a)>f(x_b)$ cambiaremos al vecino $x_b$, es decir, la probabilidad de cambiar en este caso es de $1$.
Si $f(x_a)<f(x_b)$, siempre que la temperatura $T$ sea distinta a $0$, la probabilidad de cambiar también será distinta a $0$. Entre más pequeña sea la temperatura $T$, más pequeña es la probabilidad de cambiar, lo que impedirá que el algoritmo se quede estancado en un mínimo local que no sea global y a su vez asegurar que no se escoja un punto que no cumpla con las necesidades del problema.
Eventualmente llegaremos al punto mínimo global y la temperatura será tan cercana a $0$ que ya no habrán cambios de puntos.*
"

**Detalles finales de la resolución del problema con IC (1 minuto)**

Mostramos como la metodología de IC cambia la metodología de SA.

"
*A partir de estos datos podemos calcular una suma ponderada para cada uno de los contenedores como sigue:
$$C_nT = \sum_{i=t_0}^{t_n−1} e^{−(t_n−i)}C(b,i)$$
Donde C(b,i) es la función que retorna la recompensa recibida en el instante $i$ en caso de haberse escogido el contenedor $b$ en ese instante. Es claro que en el término $t_n−i$ se da a entender que $C_nT$ favorece las recompensas que fueron más consistentes a lo largo del tiempo. A partir de esta función podemos cambiar la política de probabilidad de SA por la siguiente fórmula:
$$P_b(C_nT)=e^{−C_nT/L_t}$$
Donde L_t es la temperatura explicada en SA.*
